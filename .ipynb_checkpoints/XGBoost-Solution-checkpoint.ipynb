## Package and dataset imports

### Imports

```python
import numpy as np
import pandas as pd
from statistics import mean
import pickle

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold , StratifiedKFold
from classification_utilities import display_cm, display_adj_cm
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score, classification_report
from sklearn import preprocessing
from sklearn.impute import SimpleImputer
from sklearn.model_selection import LeavePGroupsOut

from sklearn.ensemble import  RandomForestClassifier, VotingClassifier
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
from sklearn.pipeline import make_pipeline, make_union
from sklearn.preprocessing import FunctionTransformer
import xgboost as xgb
from xgboost.sklearn import  XGBClassifier
from xgboost import plot_importance
```

```python
# Load data

facies_data = pd.read_csv('./datasets/facies_vectors.csv')

# Parameters

feature_names = ['GR', 'ILD_log10', 'DeltaPHI', 'PHIND', 'PE', 'NM_M', 'RELPOS']
facies_names = ['SS', 'CSiS', 'FSiS', 'SiSh', 'MS', 'WS', 'D', 'PS', 'BS']
#facies_colors = ['#F4D03F', '#F5B041','#DC7633','#6E2C00', '#1B4F72','#2E86C1', '#AED6F1', '#A569BD', '#196F3D']
facies_classes = list(zip(range(1,10), facies_names))

# Store features and labels
X = facies_data[feature_names].values 
y = facies_data['Facies'].values

# Store well labels and depths
well = facies_data['Well Name'].values
depth = facies_data['Depth'].values

# Fill 'PE' missing values with mean
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(X)
X = imp.transform(X)
```

## Data analysis and model training

### Creating a function to train and test model

```python
# Train and test a classifier

# Pass in the classifier so we can iterate over many seeds later.
def train_and_test(X_tr, y_tr, X_v, clf):
    
    # Feature normalization
    scaler = preprocessing.RobustScaler(quantile_range=(25.0, 75.0)).fit(X_tr)
    X_tr = scaler.transform(X_tr)
    X_v = scaler.transform(X_v)
    
    clf.fit(X_tr, y_tr)
    
    # Test classifier
    y_v_hat = clf.predict(X_v)
    
    return y_v_hat
```

## Loading blind dataset and making the prediction

```python
#Load testing data
test_data = pd.read_csv('./datasets/validation_data_nofacies.csv')
```

```python
X.shape, y.shape, test_data[feature_names].shape
```

```python
# Prepare test data
X_train = X
y_train = y
X_test = test_data[feature_names].values

y_test = pd.read_csv('./datasets/validation_data_with_facies_new.csv')
y_test = y_test['Facies']


# List of y_hat values for each iter
y_pred = []
# List of accuracies for each iter
accuracies = []
# Empty variabels used for pickling best model
best_acc = 0
best_prediction = []

n_iter = 100

print('.' * n_iter)
for seed in range(n_iter):
    np.random.seed(seed)

    # Train classifier  
    model = XGBClassifier(learning_rate=0.06,
                          max_depth=3,
                          min_child_weight=10,
                          n_estimators=80,
                          seed=seed,
                          colsample_bytree=0.9)

    # Make blind data
    # Train and test.
    y_test_hat = train_and_test(X_train, y_train, X_test, model)
    
    # Collect result.
    y_pred.append(y_test_hat)
    
    new_acc = f1_score(y_test, y_test_hat, average='micro')

     
    # Pickle model if acc is better than before
    if new_acc > best_acc:
        # Save model for later use
        pickle.dump(model, open('best_XGBoost_model.sav', 'wb'))
        # Update new best accuracy
        best_acc = new_acc
        # Update best prediciton
        best_prediction = y_test_hat
        
    
    # Add new_acc to list of accuracies
    accuracies.append(new_acc)
    
    print('|', end='')

print('\n----------')
print('Mean accuracy: ', mean(accuracies))
np.save('TIP160_boosted_100realizations.npy', y_pred)

```

```python
accuracies_df = pd.DataFrame(accuracies)

print(f'Description of results for all {n_iter} iterations: \n')
print(accuracies_df.describe())
print(f'\nClassification report for best prediction out of the {n_iter} iterations:\n')
print(classification_report(y_test, best_prediction))
```

### Exporting Classification report to xlsx

```python
cls_report = classification_report(y_test, best_prediction, output_dict=True)
cls_report_df = pd.DataFrame(cls_report).T

cls_report_df.to_csv('XGBoost-solution-report.xlsx')
```

### Calculating F1 score when accepting adjacent hits

```python
# Listing the given adjacent facies
adjacent_list = [
    [1, 2],
    [1, 2, 3],
    [2, 3],
    [4, 5],
    [4, 5, 6],
    [5, 6, 7, 8],
    [6, 7, 8],
    [6, 7, 8, 9],
    [7, 8, 9]
]

# Placeholder list for calculating mean acc of all 100 predicted arrays
adjacent_accuracies = []

# Looping through each of the 100 realizations in y_pred
for column in range(len(y_pred)):
    y_pred_local = y_pred[column]
    adj_accuracy = []
    # Looping through each int in the predicted y-hat array
    for i in range(y_pred_local.shape[0]):
    # Finding the correct Facies category
        for cat in range(1, 10):
            if y_test[i] == cat:
                # If the predicted y-value is one of the adjacent Facies to the actual Facies-class...
                if y_pred_local[i] in adjacent_list[cat-1]:
                    # ... Judge the prediction to be correct
                    adj_accuracy.append(1)
                else:
                    adj_accuracy.append(0)
    # Add accuracy for column to list of all adjacent accuracies
    adjacent_accuracies.append(mean(adj_accuracy))
    
print(f'Mean adjacent accuracies for {n_iter} iterations: ', mean(adjacent_accuracies))
```

## Summarizing results compared to 2016 competition

```python
mean_f1 = mean(accuracies)

# Defining the improvement in F1-score since 2016
improvement = (mean_f1 / 0.641 - 1) * 100

mean_adj_acc = mean(adjacent_accuracies)

print('The F1-micro score, based on 100 realizations of the model, is: {:.3f}.'.format(mean_f1))
print('This is a {:.2f} % improvement on the contest winner F1 score of 0.641.'.format(improvement))
print('The mean accuracy score for adjacent hits is {:.3f}%.'.format(mean_adj_acc))
```
